#kalkulator sederhana

```python
panjang = int(input("Masukan panjang dari persegi panjang [cm]:"))
lebar =int(input("Masukan lebar dari persegi panjang [cm]"))

luas = panjang * lebar

print(f"\nLuas dari persegi panjang yaitu {luas} cm^2")
```
